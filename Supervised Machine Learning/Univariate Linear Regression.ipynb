```python
# gradient descent assuming b = 0
alpha = 0.0001
w = 10
b = 100

m = data.shape[0]
i = 0
    
sumErrorW = 0
sumErrorB = 0
while True:
    xi = data["X"][i]
    yi = data["Y"][i]
    errorDiffW = ((w* xi + b) - yi)
        
    ErrorW = errorDiffW * xi
#   print(f"Erros is {Error}")
    sumErrorW += ErrorW
#     print(f"sumError is {sumError}")

    sqSumErrorW = (1/m) * sumErrorW
    print(f"sqSumError is {sqSumErrorW}")

    temp_w = w - (alpha * sqSumErrorW)
   
#   calculating b

    errorDiffB = (temp_w* xi + b) - yi
    sumErrorB += errorDiffB
    sqSumErrorB = (1/m) * sumErrorB
    
    temp_b = b - (alpha * sqSumErrorB)
    
    w = temp_w
    print(f"w is {w}")
    
    b = temp_b
    print(f"B is {b}")
    
    i += 1
    print("###################################")
    
    if sqSumErrorW <= 0 :
        break
    
print(f"w is {w}")
print(f"B is {b}")
```
